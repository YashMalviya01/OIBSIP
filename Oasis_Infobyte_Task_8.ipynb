Cleaning the dataset columns wise

```python
# Removing '+' and ',' from the 'Installs' column and converting to integer
# Ensure the 'Installs' column is treated as strings
apps['Installs'] = apps['Installs'].astype(str)

# Remove '+' and ',' and convert to integer
apps['Installs'] = apps['Installs'].str.replace(r'[+,]', '', regex=True).astype(int)
```

```python
#clean the price column
apps['Price'] = apps['Price'].str.replace('$', '', regex=False).astype(float)
```

```python
#conver last update to determine
apps['Last Updated'] = pd.to_datetime(apps['Last Updated'])
```

```python

print(apps.columns)

# Handling missing values
columns_to_check = ['Rating', 'Size', 'Current Ver', 'Android Ver']
# Adjust column names if needed
missing_cols = [col for col in columns_to_check if col not in apps.columns]

if missing_cols:
    print(f"Columns not found: {missing_cols}")
else:
    apps.dropna(subset=['Rating'], inplace=True)
    apps.dropna(subset=['Size'], inplace=True)
    apps.dropna(subset=['Current Ver'], inplace=True)
    apps.dropna(subset=['Android Ver'], inplace=True)
```

```python
print(User_reviews.columns)

# Adjust the column names based on the output
# Example code assuming columns might have been misnamed
User_reviews['translated_review'].fillna('', inplace=True) if 'translated_review' in User_reviews.columns else None
User_reviews['sentiment'].fillna('', inplace=True) if 'sentiment' in User_reviews.columns else None
User_reviews['sentiment_polarity'].fillna(0.0, inplace=True) if 'sentiment_polarity' in User_reviews.columns else None
User_reviews['sentiment_subjectivity'].fillna(0.0, inplace=True) if 'sentiment_subjectivity' in User_reviews.columns else None
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure the 'Price' column is numeric
apps['Price'] = pd.to_numeric(apps['Price'], errors='coerce')

# Drop rows where 'Price' is NaN
apps = apps.dropna(subset=['Price'])

# Plot distribution of app prices
plt.figure(figsize=(10, 6))
sns.histplot(data=apps, x='Price', bins=20, kde=True)
plt.title('Distribution of App Prices')
plt.xlabel('Price ($)')
plt.ylabel('Number of Apps')
plt.show()
```
